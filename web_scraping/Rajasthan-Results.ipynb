```python
from bs4 import BeautifulSoup
import requests 
```

```python
url = 'https://results.eci.gov.in/PcResultGenJune2024/partywiseresult-S06.htm'
page = requests.get(url)
if page.status_code == 200:
    soup = BeautifulSoup(page.text, 'html.parser')
    print(soup)

else:
    print(f"Failed to retrieve the page. Status code: {page.status_code}")
```

```python
table = soup.find('table', class_ ="table")
print(table)
```

```python
elec_titles = table.find_all('thead')
headers = [header.text.split() for header in elec_titles]
headers = headers[0]
print(headers)
```

```python
import pandas as pd
df = pd.DataFrame(columns = headers)

df
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data[1:-1]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
